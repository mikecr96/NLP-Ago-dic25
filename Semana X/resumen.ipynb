## Obtener el lenguaje en que está escrita la página

```python
lang_html = soup.find("html")["lang"]
```

```python
lang_html
```

```python
lang = "spanish" if "es" in lang_html else "english"
```

```python
import re
import nltk

def limpiar_texto(texto, idioma="spanish", as_text=False):
    #Lista de stopwords del idioma indicado
    stopwords = nltk.corpus.stopwords.words(idioma)
    #Convertimos todo el texto a minúsculas
    texto = texto.lower()
    #Caracteres no alfanuméricos por un espacio
    texto = re.sub(r"\W+", " ", texto)
    #Tokenizamos
    tokens = nltk.word_tokenize(texto, idioma)
    # print(f"Antes de limpiar el texto había {len(tokens)} tokens")
    #Quitamos las stopwords para quedarnos solo con palabras de contenido
    filteredTokens = [w for w in tokens if w not in stopwords]
    # print(f"Después de limpiar el texto hay {len(filteredTokens)} tokens")
    #devolvemos como lista de tokens o como string
    if as_text:
        return " ".join(filteredTokens)
    else:
        return filteredTokens
```

```python
texto_limpio = limpiar_texto(texto_completo, lang, True)
tokens_limpios = limpiar_texto(texto_completo, lang)
```

```python
contenido[:10]
```

```python
frecuencias = nltk.FreqDist(tokens_limpios)
```

```python
frecuencias.most_common(10)
```

```python
frecuencias.most_common(1)[0][1]
```

```python
max_freq = frecuencias.most_common(1)[0][1]
freq_norm = {}

for palabra, repeticiones in frecuencias.items():
    freq_norm[palabra] = repeticiones / max_freq
```

```python
freq_norm
```

```python
calif_oracion = {}

for n, p in enumerate(contenido):
    palabras_in_p = limpiar_texto(p, lang)
    # En caso de que no haya ningún token luego de la limpieza, sigo
    if len(palabras_in_p) == 0: continue
    score = 0
    for palabra in palabras_in_p:
        if palabra in freq_norm.keys():
            score += freq_norm[palabra]

    promedio = score / len(palabras_in_p)
    calif_oracion[n] = promedio
```
